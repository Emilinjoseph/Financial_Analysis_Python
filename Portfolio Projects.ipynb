# Financial Data Analysis

Before starting analysis we have to import corresponding packages. In this analysis we need four packages numpy,pandas,matplotlib and seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Next step is to read our dataset.For that first we have to check the format of dataset. Our dataset is in Excel because extension is xlsx. To read an Excel dataset we have to use read_excel function of pandas. After reading we will store it in a variable.

```python
data=pd.read_excel("E:\Powerbi\Financial Sample.xlsx")
data
```

```python
data.info()
```

Using Info function we will get more details of this dataset. This dataset contains 700 entries and 16 columns. You also get the datatype of each column and how many null values are there in each column. From the above result you can see except Discount Band column, all other columns have no null values . For the discount band its showing 647 non-null values that means rest of the values are null values.

```python
data=data.fillna(value={'Discount Band':"NoDiscount"})
data
```

Using fillna function we filled the null values with "NoDiscount". So all the null values changed to NoDiscount.Then we have to assign the new dataset to "data", otherwise it will show the old dataset with null values.

```python
data.duplicated().sum()
```

The above line is to check whether there is any duplicate rows in the dataset. To check duplicates we have to use a function named duplicated. The duplicated() method returns a Boolean Series object, with True values for the duplicate rows .We can count the number of duplicate rows by applying the sum() method to the Boolean Series object returned by the duplicated() method. So here you can see there is no duplicates in this dataset because we got zero. 

```python
data[["Units Sold","Manufacturing Price","Sale Price","Gross Sales","Discounts"," Sales","COGS","Profit"]].describe()
```

```python
print(" Net Profit in this dataset is ${}M ".format((data["Profit"].sum()/1000000).round(2)))
print(" Total Sales in this dataset is ${}M ".format((data[" Sales"].sum()/1000000).round(2)))
print(" Total Discounts in this dataset is ${}M ".format((data["Discounts"].sum()/1000000).round(2)))
```

## Country wise Analysis

```python
figure = plt.figure(figsize =(10, 3))
df=data.groupby("Country")[" Sales"].sum()
plt.pie(df,labels=df.index,shadow="True",explode = [0, 0, 0, 0,0.2],autopct='%1.1f%%')
plt.legend(title = "Country",bbox_to_anchor=(1.2, 1.2),loc='upper left',shadow="True")
plt.title("Sales by Country")
plt.show() 
```

United States of America is the country with more sales.

But when we look into the profit , France is having the most profit. Least profit we are getting from Mexico.

```python
figure = plt.figure(figsize =(10, 3))
df=data.groupby("Country")["Profit"].sum()
plt.pie(df,labels=df.index,shadow="True",explode = [0, 0.2, 0, 0,0],autopct='%1.1f%%')
plt.legend(title = "Country",bbox_to_anchor=(1.2, 1.2),loc='upper left',shadow="True")
plt.title("Profit by Country")
plt.show() 
```

## Product wise profit

Paseo is the most profitable product. Total profit we got from this product is 4.80M. Least profitable product is Carretera with total profit of 1.83M.

```python
figure = plt.figure(figsize =(10, 3))
df=data.groupby("Product")["Profit"].sum().sort_values(ascending=False)
plt.bar(df.index,df,color="red",edgecolor="black",width=0.5)
plt.xlabel('Product')
plt.ylabel('Net Profit')
plt.show()

```

## Segment wise profit

Government Segment is having more profit with 1.13 M . Enterprise is the least profitable segment with -0.61M.

```python
figure = plt.figure(figsize =(10, 3))
df=data.groupby("Segment")["Profit"].sum().sort_values(ascending=False)
plt.bar(df.index,df,color="green",edgecolor="black",width=0.5)
plt.xlabel('Segment')
plt.ylabel('Net Profit')
plt.show()
```

## Discout Band Analysis

Most of the units sold under "high" discount band and Less units sold under "No discount" band.

```python
figure = plt.figure(figsize =(10, 3))
df=data.groupby("Discount Band")["Units Sold"].sum().sort_values(ascending=False)
plt.bar(df.index,df,edgecolor="black",width=0.5)

plt.xlabel('Discount Band')
plt.ylabel('Units Sold')
plt.show()
```

Medium discount band is having more sales and again "no discount" band is having least sales.

```python
figure = plt.figure(figsize =(10, 3))
df=data.groupby("Discount Band")[" Sales"].sum().sort_values(ascending=False)
plt.bar(df.index,df,color='y',edgecolor="black",width=0.5)
plt.xlabel('Discount Band')
plt.ylabel('Net Sales')
plt.show()
```

When it comes to the profit, low discount band is more profitable compared to other bands. 

```python

figure = plt.figure(figsize =(10, 3))
df=data.groupby("Discount Band")["Profit"].sum().sort_values(ascending=False)
plt.bar(df.index,df,edgecolor="black",width=0.5)
plt.xlabel('Discount Band')
plt.ylabel('Net Profit')
plt.show()
```

```python
figure = plt.figure(figsize =(10, 3))

plt.scatter(data["Discounts"],data["Month Number"],color = 'hotpink' )
plt.xlabel('Discount Band')
plt.ylabel('Net Profit')
plt.show()
```
